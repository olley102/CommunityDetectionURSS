```python
import rasterio
```

```python
with rasterio.open('/content/drive/My Drive/GoogleEarthEngine/GPW_v411_pc_time0.tif', 'r') as ds:
  time0 = ds.read()

with rasterio.open('/content/drive/My Drive/GoogleEarthEngine/GPW_v411_pc_time1.tif', 'r') as ds:
  time1 = ds.read()

print(time0.shape)
print(time1.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot()
im = ax.imshow(time0[0], vmin=0, vmax=1000)
fig.colorbar(im, ax=ax)
fig.show()
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot()
im = ax.imshow(time1[0], vmin=0, vmax=1000)
fig.colorbar(im, ax=ax)
fig.show()
```

```python
import numpy as np
```

```python
diff = np.abs(time1 - time0)
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot()
im = ax.imshow(diff[0], vmin=0, vmax=500)
fig.colorbar(im, ax=ax)
fig.show()
```
